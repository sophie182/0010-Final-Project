## This dataset has parking meter information, like the geocoordinates, zone, and more.

### Introduction
Using our dataset, I will look at two things, first the percentage of parking meters working as well as the zones with the most active parking meters. Then I will find the best neighborhood based on the amount of active parking meters. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('9ed126cc-3c06-496e-bd08-b7b6b14b4109.csv')
# Filter to include only active parking meters
active_meters = df[df['status'] == 'Active']

active_count_per_zone = active_meters['zone'].value_counts()

total_count_per_zone = df['zone'].value_counts()

percentage_active = (active_count_per_zone / total_count_per_zone) * 100
percentage_active.sort_values(inplace=True)  
plt.figure(figsize=(10, 8))  
plt.title('Percentage of Active Parking Meters by Zone') 
plt.xlabel('Zone')  
plt.ylabel('Percentage Active (%)') 
plt.xticks(rotation=90)  
plt.ylim(0, 100)
percentage_active.plot(kind='bar', width=0.8)
plt.show()  # Display the plot
```

## the graph above shows that zones 354, 355, Garage at 2nd avenue and mon wharf have the highest percentage of active parking meters. 

```python
sorted_active_count = active_count_per_zone.sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 8)) 
sorted_active_count.plot(kind='bar', width=0.8)  
plt.title('Zones with the Most Active Parking Meters') 
plt.xlabel('Zone')  
plt.ylabel('Number of Active Meters')  
plt.xticks(rotation=90)  
plt.show() 
```

this graph gives us more context on the first graph because we can now see that some of zones with the highest percentages have very little parking meters. Zone 355 has a very high percentage of active parking meters but not that many of them in the first place for example.
